```python
import pandas as pd
import numpy as np
```

```python
# Loading the Data
# What it does: Reads the Long sheet from the source file into Python, skipping the first 3 junk rows.
SourceFile = pd.read_excel('Maternal-and-Newborn-Health-Coverage-Database-December-2023.xlsx',sheet_name='Long', header=3)
print("DataFrame Loaded")
```

```python
#Initial Inspection
#Displaying the head of the DataFrame:
display(SourceFile.head())
```

```python
#Copying the sourcefile into df dataframe
df = SourceFile.copy()
display(df.head())
display(df.tail())
```

```python
#Performing some EDA 
display(df.describe())
display("No of rows and columns: ",df.shape)
display(df.info())
```

```python
# Data Cleaning and Filtering
# For the indicators tab, we are considering only the indicators that are relevant to our research hypothesis.
# The indicators of interest are:
# - ANC1: Antenatal care 1+ visit
# - ANC4: Antenatal care 4+ visits
# - SAB: Skilled birth attendance
# - INSTDEL: Institutional deliveries
# - PNCMOM: Postnatal care for mothers
# - PNCNB: Postnatal care for newborns
# - Demand_FP: Family planning demand satisfied

#filtering the dataframe to only include the indicators from above:
indicators = ['ANC1', 'ANC4', 'SAB', 'INSTDEL', 'PNCMOM', 'PNCNB', 'DEMAND_FP']
df = df[df['Indicator'].isin(indicators)]
display(df.shape)
```

```python
#Drop the unnecessary column: In this step, we will drop the unnecessary column for our project:
columns_to_drop = ["UNICEF Reporting Region",
                   "UNICEF Programme Region",
                   "World Bank Income Group (2023)",
                   "Year(s) of data collection",
                   "Short Source","Long Source",
                   "Sample Size Notes",
                   "Notes",
                   "Source Notes"]
df_modified = df.drop(columns=columns_to_drop, errors='ignore')
df = df_modified.copy()
display(df.head())
```

```python
#counting the data for the relevant column
all_unique_counts = {
    "Indicator": len(df["Indicator"].unique()),
    "ISO Code": len(df["ISO Code"].unique()),
    "Countries and areas": len(df["Countries and areas"].unique()),
    "Stratifier": len(df["Stratifier"].unique()),
    "Level": len(df["Level"].unique()),
    "Value": len(df["Value"].unique()),
    "Latest Year": len(df["Latest Year"].unique())
}

print(all_unique_counts)
```

```python
# Now considering the Hypothesis we have we will prepare our Data. 
# H1: Women living in urban areas have higher access to maternal healthcare services than those in rural areas.
# H2: Women from wealthier household is likely to receive antenatal, delivery and postnatal care services
# H3: The discrepancy between rich and poor women is larger in rural areas than urban areas.
# Making the DataFrame to support our hypothesis.

relevant_indicators = ['ANC4', 'INSTDEL', 'PNCMOM', 'PNCNB', 'ANC1']
h_df = df[df["Indicator"].isin(relevant_indicators)]

#for hypotheses 1 and 2, let's consider the Statifier in Area and WIQ
h1_df = h_df[h_df["Stratifier"].isin(['Area', 'WIQ'])]
display(h_df.head())
display(h1_df.head())
```

```python
# Handling if there are any missing values
# Remove the rows with:
# - Missing Values
# - Missing Level
# - Value outside of 0-100
h2_df = h1_df.dropna(subset=['Value','Level'])   #only cheking Value and level column for missing values
h2_df = h1_df[(h1_df["Value"] >= 0) & (h1_df["Value"] <= 100)]
display(h2_df)
```

```python
#Further Exploratory Data Analysis
df = h2_df.copy()
display(df.head())
display(df.tail())
display(df.info())
display(df.shape)

#Unique indicators
print("\n")
display(df["Indicator"].value_counts())

# Stratifiers in our filered dataset:
print("\n")
display(df["Stratifier"].value_counts())

# The level in each of the Stratifier and their counts
print("\n")
display(df.groupby("Stratifier")["Level"].value_counts())


#Checking value distribution to ensure there are no abnormal values in the Value column.
print("\nValue distribution according to stratifier and their levels:")
display(df["Value"].describe())

# Checking Year Distribution. How many observations per year?
print("\n")
display(df["Latest Year"].value_counts().sort_index())

# making sure the is no Null values:
print("\n")
display(df.isna().sum())
```

```python
#Creating sub tables
#Creating two dataframe with filter for both area and wealth
df_area = df[df["Stratifier"] == "Area"].copy()
df_wealth = df[df["Stratifier"] == "WIQ"].copy()
df_area = df_area.reset_index()
df_wealth = df_wealth.reset_index()
display(df_area.info())
display(df_wealth.info())
```

```python
# category count for both area and wealth df 
display(df_area["Level"].value_counts())
display(df_wealth["Level"].value_counts())
```

```python
# Hypothesis testing
# Hypothesis 1: Urban vs Rural Access (Calculating Urban Rural Gap for Maternal Health Access)
df_area_pivot = (df_area.pivot_table(index = ["ISO Code", "Countries and areas", "Latest Year", "Indicator"], 
                                     columns="Level", 
                                     values="Value").reset_index())
df_area_pivot["Urban_Rural_Gap"] = df_area_pivot["Urban"] - df_area_pivot["Rural"]

display(df_area_pivot)
```

```python
# Hypothesis 2: Wealth Gradient (Rich and Poor Gap for Maternal Health Access)
df_wealth_pivot = (df_wealth.pivot_table(index = ["ISO Code","Countries and areas", "Latest Year","Indicator"], 
                                         columns="Level",
                                         values="Value").reset_index())
df_wealth_pivot["Rich_Poor_Gap"] = df_wealth_pivot["Richest"] - df_wealth_pivot["Poorest"]
display(df_wealth_pivot)
```

```python
# Hypothesis 3: The discrepancy between rich and poor women is larger in rural areas than urban areas.
# What it does: Merges the H1 table and H2 table together into one master dataframe (df_merge).
# Why: To test Hypothesis 3 (Systemic Inequality), I need to see if the same countries that fail rural women also fail poor women. You need both gaps in one row to do this.
# We have to merge df_area and df_wealth to get a combination of Area(Urban/Rural and Wealth(Poorest->Richest) , Value(Coverage%))
df_merge = df_area.merge( df_wealth, on=["ISO Code","Countries and areas","Latest Year","Indicator"], 
                         how="inner",
                         suffixes=("_area","_wealth"))
display(df_merge)
```

```python
#Pivot so Urban/Rural get their own quantiles
df_h3 = df_merge.pivot_table(index=["ISO Code", "Countries and areas", "Latest Year", "Indicator", "Level_area"],
                             columns="Level_wealth",
                             values="Value_wealth")
df_h3.columns.name = None #Removes the axis name
df_h3 = df_h3.reset_index()

#Compute the wealth gap for Rich Poor Area
df_h3["Rich_Poor_Gap"] = df_h3["Richest"] - df_h3["Poorest"]
display(df_h3)
df_rich_poor = df_h3.copy()
```

```python
#Now Creating a comparison table for rich poor gap for Urban and Rural area
# Create a Comparison Table (Final Output for H3)
df_gap_compare = df_h3.pivot_table(index=["ISO Code", "Countries and areas", "Latest Year", "Indicator"],
                                  columns = "Level_area",
                                  values = "Rich_Poor_Gap" ).reset_index()
display(df_gap_compare)
```

âš ï¸ Data Limitation Note:  We cannot calculate "Rural Poverty" directly

It is important to note that df_merge combines two independent stratifications:

    Area Stratification: Shows the average coverage for all Urban vs. all Rural women.

    Wealth Stratification: Shows the average coverage for all Poorest vs. all Richest women.

The Limitation: The dataset does not provide nested (cross-tabulated) data.

    We have data for "Urban Women" and "Poorest Women."

    We do not have data for the specific intersection of "Poorest Women living in Urban areas."

    Example: The value 44.0 for "Poorest" in the Afghanistan row is the national average for poor women, regardless of where they live.

The Solution: Because we cannot calculate the intra-rural wealth gap directly, we instead use Correlation Analysis (Hypothesis 3). We test if countries with high Geographic Inequality also tend to have high Economic Inequality, thereby identifying systemic failure.

```python
# ---------------------------------------------------------
# DATA LIMITATION NOTE:
# This dataframe merges 'Area' stats and 'Wealth' stats side-by-side.
# IMPORTANT: These are INDEPENDENT variables, not nested.
# We CANNOT calculate "Urban-Poor" vs "Rural-Poor" from this table.
# ---------------------------------------------------------
#Now Calculating Urban-Rural gap for Hypotheis 3 (Urbanâ€“Rural_Gap = Value_Urban â€‹âˆ’ Value_Ruralâ€‹)
#Pivoting the df_merge table for this hypothesis.
df_urban_rural = df_merge.pivot_table(index = ["ISO Code","Countries and areas","Latest Year","Indicator"],
                                      columns = "Level_area",
                                      values = "Value_area").reset_index()

#create a new column for Urban-Rural Gap
df_urban_rural["Urban_Rural_Gap"] = ( df_urban_rural["Urban"] - df_urban_rural["Rural"] )
display(df_urban_rural)
df_urban_rural_gap = df_urban_rural.copy()
```

```python
# 1. H1 Visualization: The Dumbbell Chart. This chart is superior to a bar chart because it clearly shows the distance (inequality) between Urban and Rural coverage for the most unequal countries.
import matplotlib.pyplot as plt
import seaborn as sns

# Function to draw a Dumbbell Chart
def plot_dumbbell(df, indicator, gap_col, col_low, col_high, title, color_low='red', color_high='blue'):
    # Filter for the indicator and take top 15 gaps
    data = df[df['Indicator'] == indicator].sort_values(by=gap_col, ascending=False).head(10)
    
    plt.figure(figsize=(10, 8))
    
    # 1. Draw the lines connecting the dots
    plt.hlines(y=data['Countries and areas'], xmin=data[col_low], xmax=data[col_high], 
               color='grey', alpha=0.4, linewidth=3)
    
    # 2. Draw the dots
    plt.scatter(data[col_low], data['Countries and areas'], color=color_low, alpha=1, s=100, label=col_low)
    plt.scatter(data[col_high], data['Countries and areas'], color=color_high, alpha=1, s=100, label=col_high)
    
    # Formatting
    plt.title(title, fontsize=14)
    plt.xlabel('Coverage (%)')
    plt.legend()
    plt.grid(axis='x', alpha=0.3)
    plt.tight_layout()
    plt.show()

# Run the function for H1 (Urban vs Rural)
# Ensure your df_area_pivot is ready from your previous cells
plot_dumbbell(df_area_pivot, 'ANC4', 'Urban_Rural_Gap', 'Rural', 'Urban', 
              'Top 10 Countries by Urban-Rural Gap (Antenatal care 4+ Doctor Visit before Delivery)', 
              color_low='#d95f02', color_high='#1b9e77')
plot_dumbbell(df_area_pivot, 'INSTDEL', 'Urban_Rural_Gap', 'Rural', 'Urban', 
              'Top 10 Countries by Urban-Rural Gap (Institutional Delivery)', 
              color_low='#d95f02', color_high='#1b9e77')

# Run the function for H2 (Rich vs Poor)
# Ensure your df_wealth_pivot is ready
plot_dumbbell(df_wealth_pivot, 'ANC4', 'Rich_Poor_Gap', 'Poorest', 'Richest', 
              'Top 10 Countries by Rich-Poor Gap (Antenatal care 4+ Doctor Visit before Delivery)', 
              color_low='#7570b3', color_high='#e7298a')
plot_dumbbell(df_wealth_pivot, 'INSTDEL', 'Rich_Poor_Gap', 'Poorest', 'Richest', 
              'Top 10 Countries by Rich-Poor Gap (Institutional Delivery)', 
              color_low='#7570b3', color_high='#e7298a')
```

```python
#"To test our revised Hypothesis 3, I merged the geographic and economic disparity datasets to create a unified view of inequality for each country. 
# I then performed a Pearson Correlation Analysis. This statistical test determines if there is a significant relationship between a country's Urban-Rural 
# gap and its Rich-Poor gap. A strong positive correlation would confirm that these inequalities are reinforcing rather than independent."

import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

# --- Step 1: Prepare the H3 Data ---
# Merge the Urban-Rural Pivot with the Wealth Pivot
# We join on Country, Year, and Indicator to ensure we are comparing the exact same survey data
df_h3_analysis = pd.merge(
    df_area_pivot[['ISO Code', 'Countries and areas', 'Latest Year', 'Indicator', 'Urban_Rural_Gap']],
    df_wealth_pivot[['ISO Code', 'Countries and areas', 'Latest Year', 'Indicator', 'Rich_Poor_Gap']],
    on=['ISO Code', 'Countries and areas', 'Latest Year', 'Indicator'],
    how='inner'
)

# --- Step 2: Statistical Test (Pearson Correlation) ---
print("--- H3 Statistical Test Results (Correlation Analysis) ---\n")

# We will test this for each indicator separately
indicators_list = df_h3_analysis['Indicator'].unique()

for ind in indicators_list:
    subset = df_h3_analysis[df_h3_analysis['Indicator'] == ind]
    
    # Drop missing values to ensure accurate stats
    subset = subset.dropna(subset=['Urban_Rural_Gap', 'Rich_Poor_Gap'])
    
    if len(subset) > 10:  # Only test if we have enough countries
        # Calculate Pearson Correlation
        corr_coef, p_value = stats.pearsonr(subset['Urban_Rural_Gap'], subset['Rich_Poor_Gap'])
        
        print(f"Indicator: {ind}")
        print(f"  Correlation Coefficient (r): {corr_coef:.3f}")
        print(f"  P-value: {p_value:.5f}")
        
        if p_value < 0.05 and corr_coef > 0:
            print("   CONCLUSION: Statistically significant POSITIVE correlation.")
            print("    (Countries with larger Urban-Rural gaps also have larger Rich-Poor gaps)")
        else:
            print("   CONCLUSION: No significant correlation found.")
        print("-" * 50)

# --- Step 3: Visualization (Scatter Plot with Regression Line) ---
# We will plot 'Institutional Delivery' (INSTDEL) as the main example
target_indicator = 'INSTDEL'
plot_data = df_h3_analysis[df_h3_analysis['Indicator'] == target_indicator]

plt.figure(figsize=(10, 6))
sns.regplot(data=plot_data, x='Urban_Rural_Gap', y='Rich_Poor_Gap', 
            scatter_kws={'s': 80, 'alpha': 0.6, 'color': '#2b8cbe'},
            line_kws={'color': '#e34a33', 'linewidth': 2})

plt.title(f'H3: Systemic Inequality Analysis ({target_indicator})\n(Correlation between Geography and Wealth)', fontsize=14)
plt.xlabel('Urban - Rural Gap (Geographic Penalty)', fontsize=12)
plt.ylabel('Rich - Poor Gap (Wealth Penalty)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)

# Label the top 3 most unequal countries on the plot
top_countries = plot_data.sort_values(by='Rich_Poor_Gap', ascending=False).head(3)
for i, row in top_countries.iterrows():
    plt.text(row['Urban_Rural_Gap']+1, row['Rich_Poor_Gap'], row['Countries and areas'], fontsize=9, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
import os
os.environ["OMP_NUM_THREADS"] = "1"
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import seaborn as sns
import matplotlib.pyplot as plt

# --- 1. Prepare Data for ML ---
# We will focus on one key indicator for clustering, e.g., 'INSTDEL' (Institutional Delivery)
# You can change this to 'ANC4' or others to see different patterns
target_indicator = 'INSTDEL'
df_ml = df_h3_analysis[df_h3_analysis['Indicator'] == target_indicator].copy()

# Drop rows with missing values to avoid errors
df_ml = df_ml.dropna(subset=['Urban_Rural_Gap', 'Rich_Poor_Gap'])

# Select features for clustering
features = df_ml[['Urban_Rural_Gap', 'Rich_Poor_Gap']]

# --- 2. Standardize the Data ---
# ML algorithms need data on the same scale
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# --- 3. Run K-Means Clustering ---
# We choose k=3 to find 3 distinct groups of countries
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df_ml['Cluster'] = kmeans.fit_predict(features_scaled)
#making a backup of df_ml
df_final = df_ml.copy()
# Map cluster numbers to meaningful names (You may need to adjust these based on the plot output!)
# We sort clusters by their average "Rich_Poor_Gap" to try and order them consistently
cluster_means = df_ml.groupby('Cluster')['Rich_Poor_Gap'].mean().sort_values()
# Create a dictionary to rename clusters: 0 -> Low, 1 -> Med, 2 -> High (roughly)
cluster_map = {
    cluster_means.index[0]: 'Low Inequality (Best)',
    cluster_means.index[1]: 'Moderate Inequality',
    cluster_means.index[2]: 'High Inequality (Worst)'
}
df_ml['Cluster_Label'] = df_ml['Cluster'].map(cluster_map)

# --- 4. Visualize the Clusters ---
plt.figure(figsize=(10, 7))
sns.scatterplot(data=df_ml, x='Urban_Rural_Gap', y='Rich_Poor_Gap', 
                hue='Cluster_Label', palette='viridis', s=100, edgecolor='black', alpha=0.8)

# Add centroids (centers of the clusters)
centroids = scaler.inverse_transform(kmeans.cluster_centers_)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=300, marker='X', label='Centroids')

# Label a few countries to make it interesting
for i, row in df_ml.iterrows():
    if row['Rich_Poor_Gap'] > 50 or row['Urban_Rural_Gap'] > 40: # Label extreme cases
        plt.text(row['Urban_Rural_Gap']+1, row['Rich_Poor_Gap'], row['ISO Code'], fontsize=9)

plt.title(f'ML Analysis: Clusters of Inequality ({target_indicator})', fontsize=15)
plt.xlabel('Urban - Rural Gap', fontsize=12)
plt.ylabel('Rich - Poor Gap', fontsize=12)
plt.legend(title='Country Cluster')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

# --- 5. Print the Countries in the "Worst" Cluster ---
worst_cluster = cluster_map[cluster_means.index[2]]
print(f"--- Countries in the '{worst_cluster}' Cluster ---(Repeatation Due to having Multiple records from different years)")
print(df_ml[df_ml['Cluster_Label'] == worst_cluster]['Countries and areas'].values)
```

```python
import warnings
warnings.filterwarnings('ignore')
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# --- 1. Prepare the Data ---
# Ensure df_ml contains your Urban_Rural_Gap and Rich_Poor_Gap
# We use the same data we used for the clustering step
X_perf = df_ml[['Urban_Rural_Gap', 'Rich_Poor_Gap']]

# --- 2. Calculate Silhouette Score (Quality Check) ---
# This measures how similar an object is to its own cluster compared to other clusters.
kmeans_model = KMeans(n_clusters=3, random_state=42, n_init=10)
labels = kmeans_model.fit_predict(X_perf)
score = silhouette_score(X_perf, labels)

print(f"Model Performance (Silhouette Score): {score:.3f}")
if score > 0.5:
    print(" Conclusion: Strong structure! The clusters are distinct and well-separated.")
elif score > 0.25:
    print(" Conclusion: Moderate structure. Some overlap exists between groups.")
else:
    print(" Conclusion: Weak structure. The data doesn't naturally separate into these groups.")

# --- 3. The Elbow Method (Validating k=3) ---
# We calculate 'Inertia' for k=1 to k=10 to show why 3 is the "sweet spot"
inertia = []
k_range = range(1, 11)

for k in k_range:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    km.fit(X_perf)
    inertia.append(km.inertia_)

# Plotting the Elbow Curve
plt.figure(figsize=(8, 5))
plt.plot(k_range, inertia, marker='o', linestyle='--', color='#2b8cbe', linewidth=2)
plt.title('The Elbow Method: Validating the Optimal Number of Clusters', fontsize=14)
plt.xlabel('Number of Clusters (k)', fontsize=12)
plt.ylabel('Inertia (Sum of Squared Errors)', fontsize=12)
plt.xticks(k_range)
plt.axvline(x=3, color='red', linestyle='--', label='Selected k=3') # Mark your choice
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
#Finally lets identify the "influence" of Wealth vs Residence using Random Forest Classifier.
# Indicator: INSTDEL (Institutional Delivery)
# Target (y): 'Access_Status' (1 = Good Access, 0 = Poor Access).
# Features (X): User Identity (Urban/Rural/Rich/Poor) + National Income.
# Outcome: We will generate a "Feature Importance" plot to show which
# factor is the dominant driver of healthcare inequality.
# ---------------------------------------------------------

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

#  Random Forest Classification (Predicting Access) ---

# 1. Prepare the Data
# Ensure 'SourceFile' is loaded (from the top of your notebook)
df = SourceFile.copy()

# Filter for Institutional Delivery (Proxy for system strength)
target_indicator = 'INSTDEL'
df_rf = df[df['Indicator'] == target_indicator].copy()

# Keep only Area and Wealth rows
df_rf = df_rf[df_rf['Stratifier'].isin(['Area', 'WIQ'])].copy()

# Clean Data
df_rf['Value'] = pd.to_numeric(df_rf['Value'], errors='coerce')
df_rf.dropna(subset=['Value'], inplace=True)
df_rf['Latest Year'] = pd.to_numeric(df_rf['Latest Year'], errors='coerce')
# Keep latest year only
df_rf = df_rf.loc[df_rf.groupby(['ISO Code', 'Stratifier', 'Level'])['Latest Year'].idxmax()]


# 2. Define Target and Features
# Target: Good Access (>= 80%)
df_rf['Access_Status'] = (df_rf['Value'] >= 80).astype(int)
y = df_rf['Access_Status']

# Features: Wealth/Residence Level & World Bank Income
features = df_rf[['Level', 'World Bank Income Group (2023)']]

# drop_first=False ensures we see bars for "Rural", "Urban", "Poorest", etc.
X = pd.get_dummies(features, drop_first=False)

# 3. Train the Model
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# 4. Results & Visualization
print(f"--- Classification Results ({target_indicator}) ---")
print("   - The higher the score (longer bar), the more 'influential' that factor is.")
print("   - The top factor is the dominant driver of inequality.")
print(f" MODEL ACCURACY: {rf_model.score(X_test, y_test):.2%}")
print("-" * 100)

# Feature Importance Plot
importances = pd.Series(rf_model.feature_importances_, index=X.columns).sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=importances.values, y=importances.index, palette='viridis')
plt.title(f'(Institutional Delivery) Determinants of Access: Wealth vs. Residence vs. National Income', fontsize=14)
plt.xlabel('Influence Score (Higher = More Important)')
plt.tight_layout()
plt.show()
```
